```python
df = pd.read_pickle("../data/popularity_5000.pkl")
df.head()
```

```python
df["popularity_new"].value_counts()
```

```python
for row in df[6000:].itertuples(index=True):
    print(f"{row.Index}, {row.track_name}, {row.artist_names}")
    
    try: 
        popularity, primary_artist, is_collab, feat_artists, duration_ms, release_date = get_song_detail_from_spotify(row.spotify_id)
        df.loc[row.Index, "popularity_new"] = popularity
        df.loc[row.Index, "is_collab"] = is_collab
        df.loc[row.Index, "primary_artist"] = primary_artist
        df.loc[row.Index, "feat_artist"] = feat_artists
        df.loc[row.Index, "duration_ms"] = duration_ms
        df.loc[row.Index, "release_date_update"] = release_date
    except:
        print("notfound and move on")
filename = f"../data/popularity_full.pkl"
df.to_pickle(filename)
# to_delete = f"data/lyrics_{200}.pkl"
# if os.path.exists(to_delete):
#     os.remove(to_delete)
```
